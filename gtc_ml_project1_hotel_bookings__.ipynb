```python
def detect_iqr_outliers(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers.shape[0], lower_bound, upper_bound

adr_outliers_count, adr_lower, adr_upper = detect_iqr_outliers(df, 'adr')
lead_time_outliers_count, lead_time_lower, lead_time_upper = detect_iqr_outliers(df, 'lead_time')

print(f"Outliers in 'adr': {adr_outliers_count} records. (Thresholds: {adr_lower:.2f} to {adr_upper:.2f})")
print(f"Outliers in 'lead_time': {lead_time_outliers_count} records. (Thresholds: {lead_time_lower:.2f} to {lead_time_upper:.2f})")
```

### **Data Quality Report**

#### **Main Data Quality Issues:**

1.  **Missing Data:** The columns `company`, `agent`, `country`, and `children` have missing values.
2.  **Outliers:** The `adr` and `lead_time` columns contain a significant number of outliers, as shown by the boxplots and the IQR method.
3.  **Invalid Data:** The `adr` column contains a negative value, which is not a valid amount.

```python
df['company'].fillna(0, inplace=True)
df['agent'].fillna(0, inplace=True)
```

```python
most_frequent_country = df['country'].mode()[0]
df['country'].fillna(most_frequent_country, inplace=True)
```

```python
median_children = df['children'].median()
df['children'].fillna(median_children, inplace=True)
```
